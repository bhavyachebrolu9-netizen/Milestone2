```python
for img_name in os.listdir("dataset/images/train"):
    label_path = f"dataset/labels/train/{img_name.split('.')[0]}.txt"

    # YOLO format: class x_center y_center width height
    labels = [
        "0 0.18 0.45 0.25 0.55",  # Photo region
        "1 0.55 0.40 0.60 0.35",  # Text region
        "2 0.80 0.55 0.25 0.40"   # QR region
    ]

    with open(label_path, "w") as f:
        f.write("\n".join(labels))
```

```python
%%writefile data.yaml
train: dataset/images/train
val: dataset/images/train

nc: 3
names: ['photo_tampering', 'text_tampering', 'qr_anomaly']
```

```python
model = YOLO("yolov8n.pt")

model.train(
    data="data.yaml",
    epochs=15,
    imgsz=640,
    batch=4
)
```

```python
from ultralytics import YOLO

trained_model = YOLO("runs/detect/train2/weights/best.pt")

```

```python
import matplotlib.pyplot as plt
import cv2
import os

for img in os.listdir("dataset/images/train"):
    print("Checking:", img)

    results = trained_model(f"dataset/images/train/{img}")
    annotated_img = results[0].plot()   # correct way

    plt.figure(figsize=(5,8))
    plt.imshow(cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()
```
