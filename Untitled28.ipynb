```python
import cv2
import numpy as np
from PIL import Image
import pytesseract
import re
from ultralytics import YOLO
from google.colab import files
```

```python
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
print("Uploaded Image:", image_path)
```

```python
def preprocess_image(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5,5), 0)
    thresh = cv2.adaptiveThreshold(
        blur, 255,
        cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
        cv2.THRESH_BINARY, 11, 2
    )
    return thresh

processed_img = preprocess_image(image_path)
cv2.imwrite("processed_aadhaar.png", processed_img)

Image.open("processed_aadhaar.png")
```

```python
text = pytesseract.image_to_string(Image.open("processed_aadhaar.png"))

print("========== OCR OUTPUT ==========")
print(text)
print("================================")
```

```python
def validate_aadhaar(text):
    aadhaar_pattern = r"\b\d{4}\s\d{4}\s\d{4}\b"
    dob_pattern = r"\b\d{2}/\d{2}/\d{4}\b"

    aadhaar_valid = bool(re.search(aadhaar_pattern, text))
    dob_valid = bool(re.search(dob_pattern, text))

    return aadhaar_valid, dob_valid

aadhaar_ok, dob_ok = validate_aadhaar(text)

print("Aadhaar Number Valid:", aadhaar_ok)
print("DOB Format Valid:", dob_ok)
```

```python
def improve_text(text):
    return text.replace("0", "O")

improved_text = improve_text(text)
print("Improved OCR Text:\n", improved_text)
```

```python
model = YOLO("yolov8n.pt")
```

```python
results = model(image_path, conf=0.3)

if len(results[0].boxes) > 0:
    yolo_flag = True
    print(" YOLO: Possible forgery detected")
else:
    yolo_flag = False
    print(" YOLO: No forgery detected")
```

```python
def anomaly_score(image_path):
    img = cv2.imread(image_path, 0)
    edges = cv2.Canny(img, 100, 200)
    return np.sum(edges) / 100000

score = anomaly_score(image_path)
anomaly_flag = score > 5

print("Anomaly Score:", score)
```

```python
def final_decision(aadhaar_ok, dob_ok, yolo_flag, anomaly_flag):
    if not aadhaar_ok or yolo_flag or anomaly_flag:
        return " FRAUD DOCUMENT"
    return " VERIFIED DOCUMENT"

decision = final_decision(aadhaar_ok, dob_ok, yolo_flag, anomaly_flag)
```

```python
print("\n===================================")
print(" AI AADHAAR FRAUD DETECTION SYSTEM ")
print("===================================")
print("Aadhaar Valid :", aadhaar_ok)
print("DOB Valid     :", dob_ok)
print("YOLO Flag     :", yolo_flag)
print("Anomaly Score :", score)
print("-----------------------------------")
print("FINAL STATUS  :", decision)
print("===================================")
```
